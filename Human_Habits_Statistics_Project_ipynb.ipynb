```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
df = pd.read_csv("human_habits_variability_statistics_project.csv")
df.head()
```

```python
df.info()
df.describe()
```

```python
df.isnull().sum()
```

```python
df.std()
```

```python
df.select_dtypes(include="number").std()
```

```python
df.groupby("Day_Type").mean(numeric_only=True)
```

```python
# Probability of healthy sleep (>7 hours)
prob_sleep = len(df[df["Sleep_Hours"] > 7]) / len(df)
prob_sleep
```

```python
weekday_sleep = df[df["Day_Type"] == "Weekday"]["Sleep_Hours"]
weekend_sleep = df[df["Day_Type"] == "Weekend"]["Sleep_Hours"]

t_stat, p_value = stats.ttest_ind(weekday_sleep, weekend_sleep)

print("T-statistic:", t_stat)
print("P-value:", p_value)
```

Key Findings:

Screen time shows the highest variability, meaning it is the most unpredictable habit.

Physical activity has the lowest standard deviation, indicating consistent behavior.

Approximately 48% of individuals achieve healthy sleep duration (>7 hours).

Behavioral patterns differ between weekdays and weekends.

Statistical testing validates whether these differences are significant.

```python
# Calculate standard deviation for numeric columns
std_values = df.select_dtypes(include="number").std()

# Create bar chart
plt.figure(figsize=(8,5))
std_values.plot(kind='bar')
plt.title("Variability of Human Daily Habits (Standard Deviation)")
plt.ylabel("Standard Deviation")
plt.xlabel("Habit")
plt.xticks(rotation=0)
plt.show()
```
